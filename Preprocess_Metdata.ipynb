```python
import os
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
```

```python
from google.colab import files

# Upload a file and save its path
uploaded = files.upload()
# Access the uploaded file
for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])}')
```

```python
from IPython.display import display, HTML

# List to store dataframes
dfs = []

headers = ["Audio", "Transcript", "Gender", "Age", "Accent", "Mood", "Environment", "Duration"]

for filename in uploaded.keys():
      # Read the CSV file; skip header if not the first one
      df = pd.read_csv(filename, header=None)
      if df.iloc[0][0] == headers[0]:
          df = df[1:].reset_index(drop=True)

      # Append the dataframe
      dfs.append(df)

# Concatenate all dataframes
final_df = pd.concat(dfs, ignore_index=True)

# print(f"Calculated Final Length: {final_length}")
print(f"Actual Final Length: {len(final_df)}")
display(HTML(final_df.head(2).to_html()))

# Save the concatenated dataframe to a new CSV file
final_df.to_csv('/content/drive/My Drive/metadata.csv', index=False)

final_df = pd.read_csv('/content/drive/My Drive/metadata.csv', names=headers)
final_df = final_df.iloc[1:]
final_df.to_csv('/content/drive/My Drive/metadata.csv', index=False)
```
